 **1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.**

dog pig

pig ham

```python
# 1-1
with open("s.txt", "r") as f:
    linelist = []
    for line in f:
        linelist.append(line.rstrip('\n'))
    
    firstlist = []
    for i in range(len(linelist)):
        firstlist.append(linelist[i].split()[0])
    firstlist.sort()
    
    anslist = []
    for i in range(len(firstlist)):
        for j in range(len(linelist)):
            if firstlist[i] == linelist[j].split()[0]:
                anslist.append(linelist[j])
                
with open("s1.txt", "w") as f:
    for i in range(len(anslist)):
        f.write(anslist[i])
        if i != (len(anslist) - 1):
            f.write('\n')
```

### [3번 문제]
### range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.

* **1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.**
    range(stop)
    range(start, stop[, step])

* **2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.**
* **3) frange 사용 예**
  * 3-1) frange(0.5)
    * [0.0, 0.1, 0.2, 0.3, 0.4]
  * 3-2) frange(1.0, 2.0)
    * [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
  * 3-3) frange(2.2, 4.0, 0.5)
    * [2.2, 2.7, 3.2, 3.7]

```python
def frange(start, stop = None, step = 0.1):
    if(stop == None):
        stop = start
        start = 0.0
    anslist = []
    while(start < stop):
        start = int(start*10)/10
        anslist.append(start)
        start += step
    print(anslist)


frange(0.5)
frange(1.0, 2.0)
frange(2.2, 4.0, 0.5)
```

 //매개변수가 1개일 경우 입력 받는 매개변수는 stop이기 때문에
 
입력받은 매개변수(start)를 stop에 넣어주고 start에는 디폴트값인 0.0을 넣어준다.
 
start가 stop이랑 같거나 작은 경우 anslist에 추가하고
 
step만큼 더해준다.(step의 디폴트값은 0.1)
 
\+ 0.1 + 0.2는 정확한 0.3이 아니기 때문에 부동 소수점 숫자를 정확히 포함하기 위해 anslist에 추가하기 전 start = int(start\*10)/10를 해준다.

### [4번 문제]
### 가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오

* **sum() 사용 예**
  * sum()
    * 0
  * sum(1, 2)
    * 3
  * sum(1, 2, 3, 4, 5)
    * 15
  * sum(1, 5, 7, 2, -10)
    * 5

```python
def sum(*args):
    ans = 0
    for i in args:
        ans = ans + i
    print(ans)

sum()
sum(1, 2)
sum(1, 2, 3, 4, 5)
sum(1, 5, 7, 2, -10)
```

 //가변인수(\*args)를 매개변수로 받고
 
 모든 args를 0으로 초기화 되어있는 ans에 계속 더해주고 출력하는 sum()을 구현한다.

* **2) myinitial() 함수 사용 예**
  * myinitial("as soon as possible")

```python
s = input("문자열을 입력하시오 : ")
def myinitial(s):
    print(''.join(map(lambda s : s[0].upper(), s.split())))
myinitial(s)
```

### 음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.

* **[옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.**

```python
def myfact(n):
    if n < 0: return
    elif n == 1: return 1
    else: return n * myfact(n-1)
n = int(input("n : "))
print("n! = ", end = "")
print(myfact(n))
```

 //정수 n을 입력받고 n이 음수가 아니면 n과 myfact(n-1)를 곱하여 반환한다.
 
myfact(n-1)은 n-1과 myfact(n-2)를 곱하여 반환하고

myfact(n-2)은 n-2와 myfact(n-3)을 곱하여 반환한다.

매개변수가 1이 들어가면 1을 반환하여 결과적으로 myfact(n)은

1부터 n까지 곱한 값을 반환하게 된다.
