* 다형성은 상속 관계 내의 다른 클래스들의 인스턴스들이 같은 멤버 함수 호출에 대해 각각 다르게 반응하도록 하는 기능이다.


* 다형성이라는 기능을 이용하면
  * 적은 코딩으로 다양한 객체들에게 유사한 작업을 수행시킬 수 있으며
  * 프로그램의 작성 코드 량이 줄어들고
  * 코드의 가독성을 높혀준다.
  * 특히 파이썬은 실시간으로 객체의 형이 결정되므로 단 하나의 메소드에 의해 처리될 수 있는 객체의 종류에 제한이 없다.

```python
class vehicle():
    def speed(self):
        print("unknown")
        
class car(vehicle):
    def speed(self):
        print("The speed of the car is 80km/s")
        
class cycle(vehicle):
    def speed(self):
        print("The speed of the cycle is 20km/s")
    
class train(vehicle):
    pass

for each in (car(), cycle(), train()):
    each.speed()
```

  * \>\>\> c = c + 5 

  * \>\>\> d = d - 5 

  * \>\>\> print(c) 

  * \[Count (step: 1)\] 17 

  * \>\>\> print(d) 

  * \[Count (step: 2)\] 9 


* **요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.**

  * \>\>\> print(c) > 10 

  * True 

  * \>\>\> print(d) > 10 

  * False 

  * \>\>\> print(c) < 10 

  * False 

  * \>\>\> print(d) < 10 

  * True 

  * \>\>\> print(c) == 17 

  * True 

  * \>\>\> print(d) != 9 

  * False 

```python
class Counter:
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
        
    def __str__(self):
        return "[Count(step:{0})] {1}".format(self.step, self.count)
    
    def incr(self):
        self.count += self.step
        
    def __call__(self):
        return self.incr()
    
    def __add__(self, num):
        self.count += num
        return self
    
    def __sub__(self, num):
        self.count -= num
        return self
    
    def __gt__(self, num):
        return self.count > num
    
    def __lt__(self, num):
        return self.count < num
    
    def __eq__(self, num):
        return self.count == num
    
    def __ne__(self, num):
        return self.count != num
#1 init
c = Counter(10)
d = Counter(10, 2)
#2 str
print(c)
print(d)
print()
#3 incr
c.incr()
d.incr()
print(c)
print(d)
print()
#4 call
c()
d()
print(c)
print(d)
print()
#5 +, -
c = c + 5
d = d - 5
print(c)
print(d)
print()
#6 >, <, ==, !=
print(c > 10)
print(d > 10)
print(c < 10)
print(d < 10)
print(c == 17)
print(d != 9)
```

 //2. \_\_str\_\_()로 Counter의 인스턴스를 출력한다. 이 때 format을 이용한다.

 //4. \_\_call\_\_()로 인스턴스 객체를 직접 호출 한다. 이 때 self.incr()을 통해 incr()메소드를 호출한 결과 값을 반환한다.
 
 //5. \_\_add\_\_(), \_\_sub\_\_()로 입력받은 인자(num)을 self.count에 더하거나 빼고 결과를 반환한다.

### 다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다.
### 다음 클래스 정의에서 __init__(), __str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.

* **MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.**

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

 //1. 생성자 메소드 \_\_init\_\_()는 객체가 생성됨과 동시에 호출되며
 
입력받은 l의 모든 요소를 self에 append 하고 중복되는 요소를 제거하기 위해 eliminate_duplicate()를 호출한다.

 //2. print 예약어나 str() 내장함수 호출에 대응되는 메소드인 \_\_str\_\_()는 result라는 변수에 "Myset: {", self의 요소 + " ,", "}"를 추가하여 반환한다.
 
이 때 마지막 요소에서 " ,"를 제거하기 위해 "}"를 추가하기 전 슬라이싱해준다.
 
 //3. 장식자(Decorator) @staticmethod를 활용한 정적메소드 eliminate_duplicate()는 인스턴스 객체와 무관하게 클래스 이름공간에 존재하는 메소드이며,
 
클래스 이름을 이용하여 직접 호출 가능하다.

빈 리스트s에 입력받은 l의 모든 요소를 append해준다. 이 때 중복은 허용하지 않는다.

append가 끝난 후 l을 빈 리스트로 초기화 해주고 s의 요소들을 다시 l에 append해준다.

### 4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오

* **정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.**
  * 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.

    * \>\>\> u = s | t 

    * \>\>\> print(u) 

    * MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} 


  * 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.

    * \>\>\> u = s & t 
    
    * \>\>\> print(u) 

    * MySet: {2, 3} 


  * 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.

    * \>\>\> s = MySet(\[1, 2, 3\]) 

    * \>\>\> t = MySet(\[3, 4, 5\]) 

    * \>\>\> u = s - t 

    * \>\>\> print(u) 

    * MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)
            
    #1
    def __or__(self, otherset):
        s = []
        for i in self:
            s.append(i)
        for i in otherset:
            s.append(i)
            
        self.eliminate_duplicate(s)
        
        return MySet(s)
    #2
    def __and__(self, otherset):
        s = []
        for i in self:
            s.append(i)
        for i in otherset:
            s.append(i)
        
        s1 = self|otherset
        for i in s1:
            s.remove(i)
            
        return MySet(s)
    #3
    def __sub__(self, otherset):
        s = []
        for i in self:
            s.append(i)
        
        s1 = self&otherset
        for i in s1:
            s.remove(i)
            
        return MySet(s)
    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    #1
    u = s|t
    print(u)
    #2
    u = s&t
    print(u)
    #3
    s = MySet([1,2,3])
    t = MySet([3,4,5])
    u = s - t
    print(u)
```

 //1. self와 otherset의 모든 요소를 빈 리스트 s에 append한다.
 
eliminate_duplicate를 이용하여 중복을 지우고 합집합을 반환한다.
 
 //2. self와 otherset의 모든 요소를 빈 리스트 s에 append한다.
 
self와 otherset의 합집합 리스트 s1의 요소들을 s에서 remove하여 교집합을 반환한다.
 
 //3. self의 요소들을 s에 append한다.
 
self와 otherset의 교집합 리스트 s1의 요소들을 s에서 remove하여 차집합을 반환한다.
